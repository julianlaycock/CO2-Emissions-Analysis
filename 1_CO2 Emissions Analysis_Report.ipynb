## Executive Summary

This analysis examines per capita CO₂ emissions for Australia, China, Germany, India, and the United States.  
The goal is to understand long‑term trends, volatility, and relative emissions levels across countries with different economic profiles.

### Key Findings
- The United States has the highest average per capita emissions (10.16 tonnes), with peaks above 22 tonnes.
- Germany and Australia show similar averages (~5.8 tonnes), though Australia is more volatile.
- China’s emissions are lower on average (1.93 tonnes) but highly volatile, reflecting rapid industrialization.
- India reached its highest recorded per capita emissions in 2024 (2.2 tonnes), marking a significant shift from historically low levels.
- Volatility scores show China and India as the most erratic, while the U.S. and Germany are more stable.


## Methodology

1. **Data Loading**  
   The dataset was imported from Our World in Data and filtered to include five countries.

2. **Cleaning**  
   Columns were renamed for clarity: `Entity → Country`, `Annual CO₂ emissions (per capita) → CO2`.

3. **Transformation**  
   - Filtered for selected countries  
   - Sorted by Country and Year  
   - Computed year‑over‑year percentage change (`pct_change()`)

4. **Analysis**  
   - Time‑series plots  
   - Volatility calculations  
   - Summary statistics using `describe()`

5. **Limitations**  
   - India only has data for 2024 in this dataset  
   - Per‑capita emissions do not reflect total national emissions

```python
# Plot Year-over-Year Change in CO₂ Emissions Per Capita

plt.figure(figsize=(14, 6))

for c in countries:
    subset = df_filtered[df_filtered["Country"] == c]
    plt.plot(subset["Year"], subset["CO2_change_pct"], label=c)

plt.title("Year-over-Year Change in CO₂ Emissions Per Capita")
plt.xlabel("Year")
plt.ylabel("Change (%)")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```
